## Theory: What, why, where Anomaly Detection?

**Anomaly detection** (aka outlier analysis) is a step in data mining that identifies the unexpected items/events in the datasets which deviate/differ from the normal data. It has two basic assumptions:
1. Anomalies only occur very rarely in the data
2. Freatures of anomalies different from normal instances signficantly

Anomalous data can indicate:
- critical incidents (ex: technical glitch)
- potential opportunities (ex: change in consumer behaviour)

One of most famous applications of Anomaly Detection is Fraud Detection, which is an unsupervised anomaly detection problem (because data is unlabelled).
